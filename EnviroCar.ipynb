```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt

r=requests.get('https://envirocar.org/api/stable')
final=r.json()
tracks=final['tracks']
sensors=final['sensors']
phenomenons=final['phenomenons']
measurements=final['measurements']
statistics=final['statistics']
announcements=final['announcements']
badges=final['badges']
schema=final['schema']
sensors_json=requests.get(sensors).json()
measurements_json=requests.get(measurements).json()
tracks_json=requests.get(tracks).json()
phenomenons_json=requests.get(phenomenons).json()
statistics_json=requests.get(statistics).json()
announcements_json=requests.get(announcements).json()
badges_json=requests.get(badges).json()
schema_json=requests.get(schema).json()
print(json.dumps(measurements_json,indent=2))
```

```python
measurements_df=pd.json_normalize(data=measurements_json,record_path=['features'])
measurements_df
```
